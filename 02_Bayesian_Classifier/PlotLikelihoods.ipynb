```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import rc
import math

import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (5, 3.5),
         'axes.labelsize': 18,
         'axes.titlesize':18,
         'xtick.labelsize':18,
         'ytick.labelsize':18}
pylab.rcParams.update(params)
plt.rcParams.update({'font.size': 14})
rc('text', usetex=True)



def pdf(x, mu, sigma):
    """
    Calculates the normal distribution's probability density 
    function (PDF).     
    """
    term1 = 1.0 / ( math.sqrt(2*np.pi) * sigma )
    term2 = np.exp( -0.5 * ( (x-mu)/sigma )**2 )
    return term1 * term2

# generating some sample data
x = np.arange(-20, 40, 0.05)

# probability density functions
pdf1 = pdf(x, mu=4, sigma=6)
pdf2 = pdf(x, mu=14, sigma=4)
plt.figure(figsize=(6,4))
# Class conditional densities (likelihoods)
plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.ylabel('$p(x/\omega_i)$',fontsize=28)
plt.xlabel('$random~variable~x$',fontsize=28)
plt.legend([r'$\omega_1$', '$\omega_2$'], loc='upper left')
plt.annotate(r'male', xy=(1,0.06), xytext=(-4,0.08),
            arrowprops={'arrowstyle': '->'}, va='center')
plt.annotate(r'female', xy=(18,0.06), xytext=(30,0.08),
            arrowprops=dict(arrowstyle="->"), va='center')
plt.savefig('classProb.pdf') 
plt.show()
```
